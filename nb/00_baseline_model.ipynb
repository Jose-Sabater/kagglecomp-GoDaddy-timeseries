# We take the latest value of mb_density and make it our prediction value

```python
train = pd.read_csv('../data/raw/train.csv')
test = pd.read_csv('../data/raw/test.csv')
```

```python
train.head()
```

```python
train.info()
```

```python
train["first_day_of_month"] = pd.to_datetime(train["first_day_of_month"])
```

```python
train = train.sort_values(by="first_day_of_month", ascending=False).reset_index()
```

```python
train.head()
```

```python
last_target = train.groupby("cfips").head(1)[["cfips", "microbusiness_density"]]
```

```python
target_dict = dict(zip(last_target["cfips"], last_target["microbusiness_density"]))
```

```python
target_dict
```

```python
test["microbusiness_density"] = test["cfips"].map(target_dict)
```

```python
test.head()
```

```python
test[["row_id", "microbusiness_density"]].to_csv("../data/submissions/submission.csv", index=False)
```
